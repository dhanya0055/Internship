```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver.get("https://archive.ics.uci.edu")
```

```python
search=driver.find_element(By.XPATH,"/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b")
search.click()
```

```python
#scraping the Dataset name
name=[]

try:

    names=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
                                        
    for i in names:

        name.append(i.text)

except NoSuchElementException:#handling no such element exception

    name.append('No details available')

time.sleep(2)

#scraping the Data type
types=[]

try:

    types=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[2]/p")
                                        
    for i in names:

        types.append(i.text)

except NoSuchElementException:#handling no such element exception

    types.append('No details available')

time.sleep(2)

#scraping the task
task=[]

try:

    task=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[3]/p")
                                        
    for i in names:

        task.append(i.text)

except NoSuchElementException:#handling no such element exception

    task.append('No details available')

time.sleep(2)
```

```python
# scraping the attribute
attribute=[]

try:

    attribute=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[4]/p")
                                        
    for i in names:

        attribute.append(i.text)

except NoSuchElementException:#handling no such element exception

    attribute.append('No details available')

time.sleep(2)

#scarping instances
instances=[]

try:

    instances=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[5]/p")
                                        
    for i in names:

        instances.append(i.text)

except NoSuchElementException:#handling no such element exception

    instances.append('No details available')

time.sleep(2)

#scraping year
year=[]

try:

    year=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[7]/p")
                                        
    for i in year:

        year.append(i.text)

except NoSuchElementException:#handling no such element exception

    year.append('No details available')

time.sleep(2)
```

```python
# scraping the no of attribute
att=[]

try:

    att=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[6]/p")
                                        
    for i in names:

        att.append(i.text)

except NoSuchElementException:#handling no such element exception

    att.append('No details available')

time.sleep(2)
```

```python
print(len(att),len(name),len(attribute),len(year),len(instances),len(task),len(types))
```

```python
runtime=[]

for i in soup.find_all('span',class_="runtime"):
     runtime.append(i.text)
runtime
```
